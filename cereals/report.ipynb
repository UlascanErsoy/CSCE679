```python
import pandas as pd

df = pd.read_csv("data.csv", index_col=False)

df.head()
```

```python
df[df["Manufacturer"] == "A"]
```

```python
numerical = ["Calories", "Protein", "Fat", "Sodium", 
                            "Fiber", "Carbohydrates", "Sugars",
                            "Potassium", "Vitamins", "Weight", "Cups"]

og_row_count = len(df.index)
print(f"Original rows: {og_row_count}")
# cleanup negative values
outed_data = [] 

for col in numerical:
    outed_data.append(df[df[col] < 0])
    df = df[df[col] >= 0]

outed_df = pd.concat(outed_data)

print(f"Cleaned rows for missing observations: {len(df.index) - og_row_count}")

# normalize by cups
```

```python
outed_df
```

```python
df[df["Type"] == "H"]
```

```python
for numeric in numerical:
    df[numeric] = df.apply(lambda row: row[numeric] * (1. / row['Cups']), axis=1)
```
