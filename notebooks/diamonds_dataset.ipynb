```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = sns.load_dataset("diamonds")

data.head()
```

# Line Charts

```python
data["carat_rounded"] = data["carat"].round(decimals=1)

data.groupby("carat_rounded")["price"]\
    .mean()\
    .plot(x="index", y="price", title="avg. price over carat", xlabel="carat weight", ylabel="avg. price")
```

# Scatter

```python
from matplotlib.colors import ListedColormap

cmap = ListedColormap(["#4169E1",  "#87CEEB","#90EE90", "#FFD700","#FFA07A"  ])
data.plot.scatter(
    x="carat",
    y="price",
    c="cut",
    colormap=cmap,
    title="Price vs Carat"
)
```

# Bar Chart

```python
data.groupby("cut")["price"]\
    .mean()\
    .plot\
    .bar(x="index", y="price", title="Avg. Price over Cut", xlabel="Cut", ylabel="avg. price")
```

# Histogram

```python
data.hist(
    column="depth",
    bins=25
    )

plt.title("Freq. of Depth")
plt.xlabel("Depth")
plt.ylabel("Frequency")
```

# Pie Chart

```python
fig = plt.figure(figsize=(10,6))
ax = data\
        .groupby("clarity")\
        .count()\
        .plot\
        .pie(y="carat",
             autopct=lambda x: f"{x:.1f}%")

ax.legend(bbox_to_anchor=(0.04, -0.05, .95, .05), ncol=4)
plt.title("Clarity Distribution")
```
